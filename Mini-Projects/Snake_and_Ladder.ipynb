```python
import random

# Dictionary representing the positions of snakes and ladders
snakes_and_ladders = {
    3 : 9,
    16 : 4,
    5 : 25,
    36 : 6,
    7 : 49,
    64 : 8,
    9: 81,
    99 : 10  
}

# Function to roll a dice
def roll_dice():
    return random.randint(1, 6)

# Function to move the player based on the dice roll
def move_player(player, dice_roll):
    player += dice_roll
    if player in snakes_and_ladders:
        new_position = snakes_and_ladders[player]
        if new_position > player:
            print("You climbed a ladder! Move to position", new_position)
        else:
            print("Oops! You got bitten by a snake! Move to position", new_position)
        return new_position
    else:
        print("Move to position", player)
        return player

# Main function to play the game
def play_snake_and_ladder():
    player1_position = 0
    player2_position = 0
    while player1_position < 100 and player2_position < 100:
        input("Player 1: Press Enter to roll the dice")
        dice_roll = roll_dice()
        print("Player 1 rolled a", dice_roll)
        player1_position = move_player(player1_position, dice_roll)
        if player1_position >= 100:
            print("Player 1 wins!")
            break

        input("Player 2: Press Enter to roll the dice")
        dice_roll = roll_dice()
        print("Player 2 rolled a", dice_roll)
        player2_position = move_player(player2_position, dice_roll)
        if player2_position >= 100:
            print("Player 2 wins!")
            break

# Start the game
play_snake_and_ladder()
```
